```python
import pandas as pd
import numpy as np

# 1. Crear una lista aleatoria de 10 enteros en el rango de 1 a 100
lista_aleatoria = np.random.randint(1, 101, 10)

# 2. Generar una serie utilizando la lista anterior con valores de Índice de 1 a 10
# - Nombre su serie "Números aleatorios"
# - Nombra tu índice "idx"
serie = pd.Series(lista_aleatoria, index=range(1, 11), name="Números aleatorios")
serie.index.name = "idx"

print("Serie 'Números aleatorios':")
print(serie)
print(f"Tipo de datos: {serie.dtype}")
print(f"Forma: {serie.shape}")
print()

# 3. Generar los Cuadrados de todos los valores de la Serie
cuadrados = serie ** 2

print("Cuadrados de la serie:")
print(cuadrados)
print()

# 4. Mostrar los 4 últimos elementos de esta serie
print("Últimos 4 elementos de los cuadrados:")
print(cuadrados.tail(4))
print()

# 5. Mostrar todos los números >500 como una lista (sin el índice)
numeros_mayores_500 = cuadrados[cuadrados > 500].tolist()

print("Números > 500 (como lista sin índice):")
print(numeros_mayores_500)
```
